## Write Text File

Pada program sebelumnya, kita hanya bisa me-read text file. Untuk me-write  file text kita perlu menambahkan argument **mode** pada function `open()`. Berikut adalah mode yang bisa kita gunakan untuk me-write file:

<br>

Mode | Keterangan
-|-
`r+` | read dan meniban konten file paling depan dengan data baru. File yang dibuka harus ada dalam directory terlebih dahulu.
`w` | mode untuk **meniban** seluruh konten file dengan data baru. Jika file yang dibuka tidak ada, maka file akan dibuat.
`a` | mode untuk menambah konten file dengan data baru. Jika file yang dibuka tidak ada, maka file akan dibuat.
`w+` | sama seperti mode `w` tetapi juga bisa melakukan read
`a+` | sama seperti mode `a` tetapi juga melakukan read

<br>

Jika argument mode tidak di-passing, function `open()` akan secara otomatis memilih mode `r` (read-only).

Coba jalankan program berikut kemudian analisis hasilnya!

r+.PNG

**Note:** mode `r` dan `r+` hanya bisa membuka file yang sudah tersedia dalam directory.

```python
f = open('/content/sample.txt', 'r+')
f.write('Belajar Python')
f.seek(0)
print(f.read())
f.close()
```

## Tugas 3

1. Cari image apapun di internet.
2. Read dan tampilkan image.
3. Cari tahu ukuran tensor-nya.
4. Lakukan operasi aritmatika apapun pada image.
5. Tampilkan image hasil operasi aritmatika.
6. Simpan image hasil operasi aritmatika dalam folder 'latihan' dengan nama 'latihan3.png'

```python
from PIL import Image
import matplotlib.pyplot as plt
import torch
import numpy as np
import os

from google.colab import files
uploaded = files.upload()

image_filename = next(iter(uploaded))

image = Image.open(image_filename)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

image_np = np.array(image)
image_tensor = torch.tensor(image_np).float()
print('Tensor Size:', image_tensor.size())

image_tensor = image_tensor + 50
image_tensor = torch.clamp(image_tensor, 0, 255)

image_result_np = image_tensor.numpy().astype(np.uint8)
image_result = Image.fromarray(image_result_np)

plt.imshow(image_result)
plt.title('Image After Arithmetic Operation')
plt.axis('off')
plt.show()

output_folder = 'latihan'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

output_file = f'{output_folder}/latihan3.png'
image_result.save(output_file)

print(f'Image hasil operasi aritmatika telah disimpan di {output_file}')
```

## Tugas 4

1. Download file dari link berikut https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz kemudian simpan dalam folder 'latihan' dengan nama 'latihan4.tar.gz'

2. Extract file tersebut kemudan simpan hasil ekstrak di folder 'latihan'.

```python
import requests
import os
import tarfile

url = 'https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz'
output_folder = 'latihan'
output_file = os.path.join(output_folder, 'latihan4.tar.gz')

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

response = requests.get(url)
with open(output_file, 'wb') as file:
    file.write(response.content)

print(f'File telah diunduh dan disimpan sebagai {output_file}')

with tarfile.open(output_file, 'r:gz') as tar:
    tar.extractall(path=output_folder)

print(f'File telah diekstrak ke folder {output_folder}')
```
